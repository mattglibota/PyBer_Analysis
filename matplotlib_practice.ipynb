```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots(figsize =(10,5))
ax.plot(x_axis,y_axis, marker='D', color='g', 
        lw = 2, label="Boston")
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,45)
ax.grid()
```

```python
plt.barh(x_axis,y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color='cyan', 
        lw = 2, label="Chicago")
ax.invert_yaxis()
ax.legend()
ax.set_ylabel("Date")
ax.set_xlabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,45)
```

```python
y_axis_larger = [6*i for i in y_axis]
```

```python
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis_larger, color='blue', 
          alpha=0.2, edgecolors='black', linewidths=2, label='Boston')
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen",
          "greenyellow", "yellow", "orange", "gold", "indianred", 
          "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
        explode=explode_values, 
        labels=x_axis, 
        autopct='%.1f%%',
        colors=colors)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen",
          "greenyellow", "yellow", "orange", "gold", "indianred", 
          "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
       labels=x_axis,
      autopct='%.1f%%',
      colors=colors,
      explode=explode_values,
      shadow=True, counterclock=False, startangle=90)

plt.show()
```
